# Introduction à Cartopy

Cartopy est un module Python qui permet de créer des visualisations cartographiques de manière simple et rapide. Il ne remplace ni le processus cartographique complet ni un système d'information géographique, mais il est utile pour créer certains types de cartes. L'avantage est la rapidité de création des cartes et le fait que le processus de création est transparent grâce à la présence de code. 

## 1. Installation de Cartopy

Cartopy étant un module Python non standard, il faut s'assurer qu'il est installé dans notre environnement Python. Si nécessaire, on peut l'installer avec:

```python
!pip install cartopy
```

*Remarque:* Cartopy nécessite que certaines librairies sont présentes sur le système (dont GEOS ou Proj). L'installation va échouer si ces dépendances ne sont pas présentes.

## 2. Importation des bibliothèques

Cartopy est généralement utilisé ensemble avec Matplotlib. Nous chargeons également `pyproj` qui doit être utilisé pour certains systèmes de coordonnées.

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

## 3. Une première carte simple

Chaque carte nécessite une projection. La carte en tant que telle sera une visualisation Matplotlib que nous pouvons créer avec `plt.subplots()`:

```python
fig, ax = plt.subplots(
    figsize = (10, 5),
    subplot_kw = { 'projection': ccrs.EckertIV() }
)

ax.set_title('Le contour des continents, projection Eckert IV')

# Ajouter les côtes
ax.coastlines()

plt.show()
```

## 4. Un peu de contenu...

Ajoutons quelques éléments supplémentaires à la carte. Cartopy permet d'ajouter certaines couches de base qui sont référencées dans `cartopy.feature`. Les géométries viennent dans ce cas de [Natural Earth](https://www.naturalearthdata.com/). Lors du premier appel, Cartopy va télécharger les données automatiquement (et afficher du coup un message correspondant).

```python
fig, ax = plt.subplots(
    figsize = (10, 5),
    subplot_kw = { 'projection': ccrs.EckertIV() }
)

ax.set_title('Carte simple du monde entier')

# Ajouter les côtes
ax.coastlines()

# Ajouter quelques couches supplémentaires
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.RIVERS)

# Ajouter les graticules
ax.gridlines()

plt.show()
```

## 5. Personnaliser la projection et le style

Chaque élément peut être personnalisé avec des paramètres supplémentaires. Ainsi, nous pouvons changer le méridien central de la projection pour mettre l'Australie un peu plus au centre, et créer une carte un peu plus réaliste.

```python
fig, ax = plt.subplots(
    figsize = (10, 5),
    subplot_kw = { 'projection': ccrs.EckertIV(central_longitude = 130) }
)

# Ajouter quelques couches supplémentaires
ax.add_feature(cfeature.OCEAN, facecolor=cfeature.COLORS['water'])
ax.add_feature(cfeature.LAND, facecolor=cfeature.COLORS['land'])
ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='#000', linewidth=0.4)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.RIVERS)

# Ajouter les graticules
ax.gridlines(draw_labels=False, color='black', linewidth=0.1)

plt.show()
```

## 6. Ajouter quelques points

Nous pouvons ajouter des points, par exemple sur la base des coordonnées pour une série de villes.

```python
# Quelques données (fictives) de température pour plusieurs villes
cities = {
    'Paris': { 'lon': 2.3312, 'lat': 48.8621, 'temp': 12 },
    'New York': { 'lon': -73.9906, 'lat': 40.7424, 'temp': 8 },
    'Tokyo': { 'lon': 139.7635, 'lat': 35.6820, 'temp': 11 },
    'Sydney': { 'lon': 151.2076, 'lat': -33.8618, 'temp': 21 },
    'Nairobi': { 'lon': 36.8307, 'lat': -1.2928, 'temp': 26 }
}

fig, ax = plt.subplots(
    figsize = (10, 5),
    subplot_kw = { 'projection': ccrs.EckertIV(central_longitude = 130) }
)

# Définir l'étendue de la carte manuellement
ax.set_global()

# Ajouter quelques couches supplémentaires
ax.add_feature(cfeature.OCEAN, facecolor=cfeature.COLORS['water'])
ax.add_feature(cfeature.LAND, facecolor=cfeature.COLORS['land'])
ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='#000', linewidth=0.4)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.RIVERS)

# Ajouter les graticules
ax.gridlines(draw_labels=False, color='black', linewidth=0.1)

# Ajouter les villes
for city, data in cities.items():
    lon, lat, temp = data['lon'], data['lat'], data['temp']
    ax.plot(lon, lat, marker='o', color='red', markersize=5, transform=ccrs.PlateCarree())
    ax.text(lon + 3, lat, f"{city}: {temp}°C", transform=ccrs.PlateCarree(), fontsize=10)

plt.show()
```

## 7. Une carte un peu moins globale...

Il est possible d'utiliser des projections locales et définir une région plus petite:

```python
# Définir la carte avec une projection UTM zone 32N
fig, ax = plt.subplots(
    figsize = (10, 5),
    subplot_kw = { 'projection': ccrs.UTM(32) }
)

# Définir l'étendue de la carte
ax.set_extent([5.84, 10.74, 45.7, 47.84], crs=ccrs.PlateCarree())

# Ajouter un peu de contenu
ax.add_feature(cfeature.LAND, facecolor=cfeature.COLORS['land'])
ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='#000', linewidth=0.4)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.RIVERS)

plt.show()
```

## Carte des stations météo

Nous pouvons maintenant faire un exemple un peu plus complet, à savoir une carte des stations météo de MeteoSuisse. Nous allons utiliser le service WMTS de Swisstopo pour la carte de base, et télécharger la liste des stations météo depuis le site Web de MeteoSuisse.

Pour pouvoir afficher les images provenant d'un service WMTS, Cartopy nécessite quelques modules supplémentaires. Assurons-nous d'avoir `scipy`, `owslib` et `Pillow` à disposition:

```python
!pip install scipy owslib Pillow
```

Chargement du [service WMTS de Swisstopo](https://www.geo.admin.ch/fr/wmts-services-et-donnees-disponibles). Nous utilisons ici le service avec projection Pseudo-Mercator (code EPSG 3857). Cette projection est utilisée par la majorité des cartes Web (ce n'est pas pour autant un bon choix...).

Nous définissons également le nom de la couche souhaitée. Pour connaître les couches disponibles, une possibilité est de configurer QGIS pour qu'il se connecte au service WMTS de Swisstopo. Voir la vidéo [«QGIS: Ajouter un service WMS ou WMTS»](https://youtu.be/voNmaJE1HE4) pour voir comment faire ça.

```python
wmts_url = 'https://wmts.geo.admin.ch/EPSG/3857/1.0.0/WMTSCapabilities.xml'
lyrname = 'ch.swisstopo.leichte-basiskarte_reliefschattierung'
```

Nous pouvons maintenant faire une première carte:

```python
# Définir la carte avec une projection Pseudo-Mercator
fig, ax = plt.subplots(
    figsize = (18, 9),
    subplot_kw = { 'projection': ccrs.epsg(3857) }
)

# Définir l'étendue de la carte
ax.set_extent([5.84, 10.74, 45.7, 47.84], crs=ccrs.PlateCarree())

# Ajouter la couche du service WMTS
ax.add_wmts(wmts_url, layer_name=lyrname, wmts_kwargs={'Time': 'current'})

plt.show()
```

Pour la localisation des stations météo, nous pouvons les trouver dans un fichier JSON disponible à l'URL suivante:
[https://www.meteoschweiz.admin.ch/product/output/measured-values/stationsTable/messnetz-automatisch/stationsTable.messnetz-automatisch.fr.json](https://www.meteoschweiz.admin.ch/product/output/measured-values/stationsTable/messnetz-automatisch/stationsTable.messnetz-automatisch.fr.json)

Il s'agit d'un fichier en format JSON qui est facile à lire avec Python. Nous allons le télécharger directement depuis le site Web de MétéoSuisse à l'aide du module Python `requests`. Nous nous assurons donc d'abord qu'il est installé:

```python
!pip install requests
```

Nous pouvons maintenant télécharger le fichier et ensuite l'interpréter avec `json.loads`. Ceci transforme le fichier JSON en structure Python (un dictionnaire).

```python
# Importer les modules nécessaires
import json
import requests

# L'URL du fichier
url_stations_meteo = 'https://www.meteoschweiz.admin.ch/product/output/measured-values/stationsTable/messnetz-automatisch/stationsTable.messnetz-automatisch.fr.json'

# Télécharger le fichier en mémoire, le contenu du fichier sera dans la variable `response`
response = requests.get(url_stations_meteo)

# Transformer le contenu du fichier JSON en dictionnaire Python
d = json.loads(response.text)
```

Il est une bonne idée d'explorer un peu le résultat. Les stations se trouvent dans une liste que nous pouvons obtenir avec `d['stations']`. Voici juste la première:

```python
stations = d['stations']
stations[0]
```

Nous avons notamment un type de station. Nous pouvons regarder quels types existent:

```python
st_types = set([st['station_type'] for st in stations])
st_types
```

Nous allons ajouter uniquement les stations météo, et non pas les stations de mesure des précipitations. Nous pouvons maintenant faire la carte et ajouter un point par station:

```python
fig, ax = plt.subplots(
    figsize = (18, 9),
    subplot_kw = { 'projection': ccrs.epsg(3857) }
)
ax.set_extent([5.84, 10.74, 45.7, 47.84], crs=ccrs.PlateCarree())
ax.add_wmts(wmts_url, layer_name=lyrname, wmts_kwargs={'Time': 'current'})

for st in stations:
    # Extraire les coordonnées en lat/long pour chaque station
    latlong = st['latlong']
    y, x = latlong

    # Nous utilisons l'identifiant comme nom, c'est plus court
    nom = st['id']

    # Ajouter le point
    ax.plot(x, y, marker='o', color='red', markersize=5, transform=ccrs.PlateCarree())
    
    # Annoter le point avec le nom
    ax.text(x, y, nom, fontsize=8, ha='right', transform=ccrs.PlateCarree())

# Et finalement définir un titre
plt.title('Stations météo de MétéoSuisse')

plt.show()
```

La carte n'est certainement pas la plus lisible. Beaucoup de noms de stations se superposent. Et ce n'est pas trivial de résoudre ce problème, car Cartopy ne donne aucune possibilité de résoudre ce problème. Nous pouvons utiliser le module `adjustText` pour au moins améliorer un peu la situation, sans pour autant résoudre le problème entièrement. Ceci va au moins faire en sorte que les étiquettes ne se chevauchent plus, mais ça ne considère pas les points.

```python
!pip install adjustText
```

```python
from adjustText import adjust_text

fig, ax = plt.subplots(
    figsize = (18, 9),
    subplot_kw = { 'projection': ccrs.epsg(3857) }
)
ax.set_extent([5.84, 10.74, 45.7, 47.84], crs=ccrs.PlateCarree())
ax.add_wmts(wmts_url, layer_name=lyrname, wmts_kwargs={'Time': 'current'})

labels = []
for st in stations:
    # Extraire les coordonnées en lat/long pour chaque station
    latlong = st['latlong']
    y, x = latlong

    # Nous utilisons l'identifiant comme nom, c'est plus court
    nom = st['id']

    # Ajouter le point
    ax.plot(x, y, marker='o', color='red', markersize=5, transform=ccrs.PlateCarree())
    
    # Annoter le point avec le nom
    lbl = ax.text(x, y, nom, fontsize=8, ha='right', transform=ccrs.PlateCarree())
    labels.append(lbl)

adjust_text(labels, ax=ax, expand_points=(1.2, 1.2), expand_text=(1.2, 1.2), arrowprops=dict(arrowstyle='->', color='gray'))

# Et finalement définir un titre
plt.title('Stations météo de MétéoSuisse')

plt.show()
```
